## 3.reshape and resize

```python
example=np.array([[2,4,6],[4,6,3]])
print(id(example))
id(np.reshape(example,(3,2)))
```

```python
example=np.array([[2,4,6],[4,6,3]])
print(id(example))
id(example.reshape((3,2)))
```

```python
example=np.array([[2,4,6],[4,6,3]])
print(id(example))
id(example.resize((3,2)))
```

```python
example=np.array([[2,4,6],[4,6,3]])
print(id(example))
id(np.resize(example,(3,2)))
```

| function | whether returns a value | whether modify the original array | whether has the param "refcheck=False" |
| :------: | :-----: | :------: | :------: |
| A.reshape() | yes | no | no |
| np.reshape() | yes | no | no |
| np.resize() | yes | no | no |
| A.resize() | no | yes | yes |

- I am gonna test the optional param "refcheck=False"
- The truth is that "refcheck=False" is completely pointless

```python
example=np.array([[2,4,6],[4,6,3]])
example.resize(2,2)
print(example)
```

```python
example=np.array([[2,4,6],[4,6,3]])
example.resize(2,2,refcheck=False)
print(example)
```

```python
example=np.array([[2,4,6],[4,6,3]])
example.resize(3,3)
print(example)
```

```python
example=np.array([[2,4,6],[4,6,3]])
example.resize(3,3,refcheck=False)
print(example)
```
