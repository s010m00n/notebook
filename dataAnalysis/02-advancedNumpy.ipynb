## 1.sort

```python
a=np.array([[7,2,3,4],[9,12,3,5],[3,4,1,15]])
a
```

```python
#ascending
print(np.sort(a))
print(np.sort(a,axis=1))
print(np.sort(a,axis=0))
print(a)
```

```python
#ascending
print(np.argsort(a))
print(np.argsort(a,axis=1))
print(np.argsort(a,axis=0))
print(a)
```

> axis=0 means operate on column; axis=1 means operate on row

```python
#descending
a=np.array([3,2,4])#descending sort in numpy just support for vector
print(a[np.argsort(-a)])
```
